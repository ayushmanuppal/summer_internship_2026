```python
## PIE CHART - share of records per city
city_counts = df['city'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.2f%%')
plt.title('Record Share by City', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## SCATTER CHART - humidity vs rainfall
plt.figure(figsize=(10, 6))
plt.scatter(df['humidity_pct'], df['rainfall_mm'], alpha=0.5, color='#3948db')
plt.title('Humidity vs Rainfall', fontsize=14)
plt.xlabel('Humidity (%)')
plt.ylabel('Rainfall (mm)')
plt.tight_layout()
plt.show()
```

```python
## LINE CHART - temperature trend over the first 100 days
plt.figure(figsize=(12, 6))
plt.plot(df['day'][:100], df['temperature_c'][:100], color='#e74c7c')
plt.title('Temperature Trend (First 100 Days)', fontsize=14)
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.tight_layout()
plt.show()
```

```python
## HISTOGRAM - wind speed distribution
plt.figure(figsize=(10, 6))
plt.hist(df['wind_speed_kmph'], bins=20, color='#7d76b9')
plt.title('Wind Speed Distribution', fontsize=17)
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
## BAR CHART - average temperature per city
city_avg_temp = df.groupby('city')['temperature_c'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(city_avg_temp.index, city_avg_temp.values, color='#2Edc97')
plt.title('Average Temperature by City', fontsize=17)
plt.xlabel('City')
plt.ylabel('Average Temperature (C)')
plt.tight_layout()
plt.show()
```
